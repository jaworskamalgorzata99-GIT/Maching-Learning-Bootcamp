### Import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn

np.random.seed(42)
np.set_printoptions(precision=6, suppress=True)
sklearn.__version__
```

###Wygenerowanie danych

```python
pogoda = ['słonecznie', 'deszczowo', 'pochmurno', 'deszczowo', 'słonecznie', 'słonecznie',
          'pochmurno', 'pochmurno', 'słonecznie']
temperatura=['ciepło', 'zimno','ciepło','ciepło','ciepło','umiarkowanie',
             'umiarkowanie','ciepło','zimno']

spacer = ['tak','nie','tak','nie','tak','tak','nie','tak','nie']

raw_df = pd.DataFrame(data={'pogoda':pogoda,'temperatura':temperatura,'spacer':spacer})
df = raw_df.copy()
df
```

###Przygotowanie danych do modelu

```python
from sklearn.preprocessing import LabelEncoder

# Use the original raw_df to ensure columns are present
df = raw_df.copy()

# Apply LabelEncoder to 'spacer'
encoder = LabelEncoder()
df['spacer'] = encoder.fit_transform(df['spacer'])

# Apply one-hot encoding to 'pogoda' and 'temperatura'
df = pd.get_dummies(df, columns=['pogoda','temperatura'], drop_first=True,  dtype=int)
df
```

```python
data = df.copy()
target = data.pop('spacer')

data
```

```python
target
```

Klasyfikator bayesowski

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(data,target)
```

```python
model.score(data, target)
```
